data imbalance

Imbalance data are common problem in machine learning classification where there are a disproportionate ratio of observation in each class. can be found in many different areas like medical diagonisis,spam filtering, and fraud detection  

why data imbalance?
because most of machine learning algorithms work best when the number of samples in each class are about equal.This is because most algorithms are designed to accuracy and reduce error 

method of data imbalance
1.change the performance metric-accuracy is not the best metric to use when evaluating imbalanced dataset as it can very misleading.the metrics can be provide better insight icludes
confusion matrix, precision,recall,F1:score

2.change the algorithm-while in every machine learning problem,it's good rule of thumb to try
 a varietyof algorithms, it can be especially beneficial with imbalance dataset
 
3.Resampling techniques
a). undersample majority class
 undersample can be defined as removing some observation of the majority class 
 undersample can be a good choice when you have a ton of data-think millions of rows
 
b).resampling techniques-oversample minority class
   oversampling can be defined as adding more copies of the minority class 
   oversampling can be good when you don't have a ton of data to work with
   
4.generate synthetic samples
  a techniques similar to upsampling is to create synthetic samples
